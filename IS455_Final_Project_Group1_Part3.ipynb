# **Number of Crashes per Year with repsect to the Primary Contributing Cause**

**This visualization will help answer the question:**                       How do the number of cases per contributing cause changes over the years?                                          

1. We have grouped the data based on the year of the accident and the primary contributing cause of the accident.
2. It is an interactive bar chart with “Year of Accident” on the X-axis and “Total Number of Cases” on the Y-axis.  






```python
new_df = df[['CRASH_RECORD_ID','PRIM_CONTRIBUTORY_CAUSE', 'CRASH_YEAR','CRASH_DATE']]
new_df.head()
```

```python
grouped_df = new_df.groupby( [ "CRASH_YEAR", "PRIM_CONTRIBUTORY_CAUSE"] )

pc_year = pd.DataFrame(grouped_df.size().reset_index(name = "Group_Count")) 

#pc_year: Primary cause vs Year
```

```python
buttons = []
i = 0
fig3 = go.Figure()
reason_list = list(pc_year['PRIM_CONTRIBUTORY_CAUSE'].unique())
reason_list.sort()
for reason_name in reason_list:
    fig3.add_trace(
        go.Bar(
            x = pc_year['CRASH_YEAR'][pc_year['PRIM_CONTRIBUTORY_CAUSE']==reason_name],
            y = pc_year['Group_Count'][pc_year['PRIM_CONTRIBUTORY_CAUSE']==reason_name],
            name = reason_name, visible = (i==0), marker_color='lightsalmon'
        )
    )

    
for reason_name in reason_list:
    args = [False] * len(reason_list)
    args[i] = True
    

    button = dict(label = reason_name,
                  method = "update",
                  args=[{"visible": args}])
    

    buttons.append(button)
    

    i+=1
    
fig3.update_layout(updatemenus=[dict(active=0,
                                    type="dropdown",
                                    buttons=buttons,
                                    x = 0,
                                    y = 1.1,
                                    xanchor = 'left',
                                    yanchor = 'bottom'),
                              ])
fig3.update_layout(
    autosize=False,
    width=1000,
    height=800,)
```

**A very simple example of how this can be used for insights:** 

One of the primary causes is “Weather”, we can select this from the dropdown and then see how weather has contributed over the years. 
This shows that the number of accidents for the years 2018 and 2019 were relatively higher than the rest of the years. We can further use this insight to investigate the Chicago weather during those years and how it was different for other years.



**An observation:**

Irrespective of the primary cause, the year 2020 has the the smaller bar in most cases. 2020 was the year when covid pandemic started, so the number of vehicles on the streets significantly decreased than before which ultimately led to less number of crashes.

# **Speed Limit Zones in Chicago**

The visualization below plots the posted speed zones in Chicago using the location's coordinates. 

First, we classified the speed limits into three categories: 

a) *Slow Zone* - Speed limit between 0 to 15 mph

b) *Medium Zone* - Speed limit between 15 to 50 mph

c) *Fast Zone* - Speed limit over 50 mph

Then, we created three separate dataframes based on these 3 speed-limit categories. The variables in these dataframes are:

a) *Latitude*

b) *Longitude*

c) *Posted Speed Limit*

Finally, using Folium and Geopy, the zones are plotted on a map of Chicago. 

```python
df.head(5)
```

```python
df['POSTED_SPEED_LIMIT'].unique()
```

```python
conditions = [
    (df['POSTED_SPEED_LIMIT'] > 0) & (df['POSTED_SPEED_LIMIT'] <= 15),
    (df['POSTED_SPEED_LIMIT'] > 15) & (df['POSTED_SPEED_LIMIT'] <= 50),
    (df['POSTED_SPEED_LIMIT'] > 50)
    ]

# create a list of the values we want to assign for each condition
values = ['SLOW ZONE', 'MEDIUM ZONE', 'FAST ZONE']

# create a new column and use np.select to assign values to it using our lists as arguments
df['SPEED_ZONE'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

```python
speed_df = df[['LATITUDE','LONGITUDE','POSTED_SPEED_LIMIT','SPEED_ZONE']]
speed_df
```

```python
speed_df_clean = speed_df.dropna()
```

```python
slow_df = speed_df_clean.loc[(speed_df_clean['SPEED_ZONE'] == 'SLOW ZONE'), ['LATITUDE','LONGITUDE', 'POSTED_SPEED_LIMIT']]
slow_df_2 = slow_df[slow_df['LATITUDE'] != 0].groupby("LATITUDE").filter(lambda x: x['LATITUDE'].size >= 50 )
slow_df_3 = slow_df_2.drop_duplicates()
slow_df_3
```

```python
medium_df = speed_df_clean.loc[(speed_df_clean['SPEED_ZONE'] == 'MEDIUM ZONE'), ['LATITUDE','LONGITUDE', 'POSTED_SPEED_LIMIT']]
medium_df_2 = medium_df[medium_df['LATITUDE'] != 0].groupby("LATITUDE").filter(lambda x: x['LATITUDE'].size >= 100 )
medium_df_3 = medium_df_2.drop_duplicates()
medium_df_3
```

```python
fast_df = speed_df_clean.loc[(speed_df_clean['SPEED_ZONE'] == 'FAST ZONE'), ['LATITUDE','LONGITUDE', 'POSTED_SPEED_LIMIT']]
fast_df_2 = fast_df[fast_df['LATITUDE'] != 0].groupby("LATITUDE").filter(lambda x: x['LATITUDE'].size >= 5 )
fast_df_3 = fast_df_2.drop_duplicates()
fast_df_3
```

```python
from geopy.geocoders import ArcGIS

nom = ArcGIS()
nom.geocode("Chicago,Illinois")
```

```python
import folium 

map = folium.Map(location=[41.884250000000065, -87.63244999999995], zoom_start=15, popup='Chicago')
```

```python
list_slow = slow_df_3.values.tolist()
list_medium = medium_df_3.values.tolist()
list_fast = fast_df_3.values.tolist()

map2 = folium.FeatureGroup(name="speed_zones")
for i in list_slow:
  map2.add_child(folium.Marker(location=[i[0], i[1]], popup="SLOW ZONE - " +str(i[2]) + "mph", icon=folium.Icon(color='green')))

for i in list_medium:
  map2.add_child(folium.Marker(location=[i[0], i[1]], popup="MEDIUM ZONE - " +str(i[2])+ "mph", icon=folium.Icon(color='orange')))

for i in list_fast:
  map2.add_child(folium.Marker(location=[i[0], i[1]], popup="FAST ZONE -" +str(i[2])+ "mph", icon=folium.Icon(color='red')))

map.add_child(map2)
```

As we can see here, every speed zone is plot according to their coordinates on the map of Chicago. Each zone is assigned a color as follows:

a) Slow Zone - Green

b) Medium Zone - Orange

c) Fast Zone - Orange

On clicking on a location pin, users can see the speed-limit zone and the speed limit for that place.

An interesting observation from this visualization is the fact that most accidents occured in medium speed-limit zones i.e speed limit between 15 to 50 mph. This can be interpreted as people rash driving at high speeds in these locations.

The visualization shows the total number of crashes by time interval and day of week (Monday to Sunday) since 2015. The lighter the color, the less number of crashes. The darker the color, the more number of crashes.

Time interval has four categories:

1) Time of Day (e.g., 12AM - 6AM)

2) Hour (e.g., 13)

3) Month (e.g., January)

4) Year (e.g., 2015)

Clicking the drop-down menu allows the interval type to be changed. Hovering over the heatmap displays the details, including day of the week, time interval type it is currently set to, and crash count.

```python
#!pip uninstall shapely
!pip install shapely --no-binary shapely

df_clean_map=df.dropna(subset=['LATITUDE', 'LONGITUDE', 'CRASH_HOUR'])
df_clean_map['LATITUDE'] = df_clean_map['LATITUDE'].apply(pd.to_numeric)
df_clean_map['LONGITUDE'] = df_clean_map["LONGITUDE"].apply(pd.to_numeric)

def getCondition(r1=0):
    df_c = df_clean_map.loc[(df_clean_map['CRASH_HOUR'] >= r1) & (df_clean_map['CRASH_HOUR'] <= r1+1)].head(100000)
    extent = [-87, -88, 41.6, 42.1]
    central_lon = np.mean(extent[:2])
    central_lat = np.mean(extent[2:])
    plt.figure(figsize=(15, 15))
    ax = plt.axes(projection=crs.AlbersEqualArea(central_lon, central_lat))
    ax.set_extent(extent)
    plt.scatter(x=df_c['LONGITUDE'], y=df_c['LATITUDE'], 
            color="brown", s=5, transform=crs.PlateCarree())
    ax.add_feature(cartopy.feature.OCEAN)
    ax.add_feature(cartopy.feature.LAND, edgecolor='black')
    ax.add_feature(cartopy.feature.LAKES, edgecolor='black')
    plt.show()

r1=list(range(0,24,1))
widgets.interact(getCondition, r1=r1)
```

This visualization looks at the frequency of crashes at various speed limits. It is interactive because when you hover over the bars the number of accidents that have occurred at that speed limit are shown.

```python
df_clean2 = df.drop(columns = ['RD_NO', 'CRASH_DATE_EST_I', 'LANE_CNT', 'HIT_AND_RUN_I', 'INTERSECTION_RELATED_I', 'NOT_RIGHT_OF_WAY_I', 'PHOTOS_TAKEN_I', 'STATEMENTS_TAKEN_I', 'DOORING_I', 'WORK_ZONE_I', 'WORK_ZONE_TYPE', 'WORKERS_PRESENT_I'])
df_clean2 = df_clean2.dropna()
df_speedlimits = pd.unique(df_clean2['POSTED_SPEED_LIMIT'])
speedlimits = df_speedlimits.tolist()
speedlimits.sort()
allspeeds = df_clean2['POSTED_SPEED_LIMIT'].tolist()
counts = []

for x in speedlimits:
    counts.append(allspeeds.count(x))
speeddict = {}
for x in speedlimits:
    for y in counts:
        speeddict[x] = y
        counts.remove(y)
        break
speeddf = pd.DataFrame(speeddict.items())
speeddf.rename(columns = {0:'Speed Limit'}, inplace = True)
speeddf.rename(columns = {1:'Number of Crashes'}, inplace = True)

fig = px.bar(speeddf, x='Speed Limit', y='Number of Crashes', color= 'Number of Crashes', title="Speed Limit vs Number of Crashes")
fig.show()
```
